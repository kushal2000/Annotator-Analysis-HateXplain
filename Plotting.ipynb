```
with open('results_individual_cnn.pkl', 'rb') as f:
    results = pickle.load(f)

f = open('results_individual_cnn.csv', 'w')
f.write('BERT One vs All\n\n')
for parameter in results:
    f.write(str(parameter)+'\n\n')
    for c1 in sorted(results[parameter].keys()):
        sent = c1
        for c2 in sorted(results[parameter][c1].keys()):
            sent += ',' + str(round(results[parameter][c1][c2]['f1'], 2))
        f.write(sent + '\n')
    f.write('\n\n')
f.close()
```

```
import numpy as np
import glob 
files = glob.glob('Results/*.pkl')

for file in reversed(files):
    results_file = open(file.replace('.pkl', '.csv').replace('Results/', 'Tables1/'), 'w')
    try:
        with open(file, 'rb') as f:
            bert_dict = pickle.load(f)
    except:
        print('fail')
    shots = bert_dict[100].keys()
    attentions = [0, 0.001, 0.01, 0.1, 1, 10, 100]
    title = 'Davidson, CNN-GRU'
    title = ''
    for att in attentions:
        title += ',Bert '+str(att)
    title = 'Training'+title+'\n'
    results_file.write(title)

    for shot in shots:
        scores = []
        stds = []
        score = 0
        seed_scores = []
        for attn in attentions:
            score = 0
            seed_scores = []
            div = len(bert_dict[attn][shot])
            for seed in bert_dict[attn][shot].keys():
                score += bert_dict[attn][shot][seed]['f1']
                seed_scores.append(bert_dict[attn][shot][seed]['f1'])
            score /= div
            stds.append(str(round(np.std(np.array(seed_scores)),2)))
            scores.append(str(round(score,2))+' ('+stds[-1]+')')
        scores = str(shot)+',' + (','.join(scores)) + '\n'
        print(scores)
        results_file.write(scores)
```
